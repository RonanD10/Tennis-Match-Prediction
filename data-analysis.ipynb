```python
# Hardest grandslam run
# Tournament level 'G' <--> Grand Slam

slam_finals = all_matches[(all_matches['tourney_level'] == 'G') & (all_matches['round'] == 7)]
slam_finals.reset_index(inplace=True, drop=True)
slam_finals.head()

slam_runs = pd.DataFrame(columns=['Year', 'Tournament', 'Champion', 'Avg Opp Elo'])
slam_runs['Year'] = slam_finals['tourney_date'].map(lambda x: int(str(x)[:4]))
slam_runs['Tournament'] = slam_finals['tourney_name']
slam_runs['Champion'] = slam_finals['winner_name']

i = 0
for tourney_id, winner_id in zip(slam_finals['tourney_id'], slam_finals['winner_id']):
    avg_elo = np.round(all_matches[(all_matches['tourney_id'] == tourney_id) & (all_matches['winner_id'] == winner_id)]['loser_elo_before'].mean(), 3)
    slam_runs.at[i, 'Avg Opp Elo'] = avg_elo
    i += 1

slam_runs.sort_values(by='Avg Opp Elo', ascending=False).head(20)
```

```python
# Top 10 peak Elos
top_elos = all_matches.groupby('winner_name')[['winner_elo_after', 'tourney_date']].max().sort_values(by='winner_elo_after', ascending=False).head(10)
top_elos['tourney_date'] = top_elos['tourney_date'].map(lambda x: int(str(x)[:4]))
top_elos.rename(columns={'winner_elo_after': 'Peak Elo', 'tourney_date': 'Year'}, inplace=True)
top_elos.index.names= ['Name']
top_elos
```

```python
# Longest career

all_players['career_length'] = 0

for i in range(len(all_players)):
    player = all_players.loc[i, 'player_id']
    first_match = all_matches[(all_matches['winner_id'] == player) | (all_matches['loser_id'] == player)]['tourney_date'].min()
    last_match = all_matches[(all_matches['winner_id'] == player) | (all_matches['loser_id'] == player)]['tourney_date'].max()
    try:
        career_length = int(str(last_match)[:4]) - int(str(first_match)[:4])
    except:
        career_length = 0
    all_players.at[i, 'career_length'] = career_length

all_players[['name', 'career_length']].sort_values(by='career_length', ascending=False).head(10)
```

```python
# Most matches played

all_players['matches_played'] = 0

for i in range(len(all_players)):
    player = all_players.loc[i, 'player_id']
    matches_played = len(all_matches[(all_matches['winner_id'] == player) | (all_matches['loser_id'] == player)])
    all_players.at[i, 'matches_played'] = matches_played

all_players[['name', 'matches_played']].sort_values(by='matches_played', ascending=False).head(10)
```
